```python
import numpy as np

# define a quantum state!
state_0 = np.array([1, 0])
state_1 = np.array([0, 1])

state_plus = np.array([1 / np.sqrt(2), 1 / np.sqrt(2)])
state_minus = np.array([1 / np.sqrt(2), -1 / np.sqrt(2)])
```

```python
#np.dot(state_0.T, state_1)
np.dot(state_plus.T, state_minus)
```

```python
xgate = np.array([[0, 1], [1, 0]])# x gate
state_0_1 = np.matmul(xgate, state_0)# change state_0 into state_1
print(state_0_1)
state_minus_1 = np.matmul(xgate, state_minus)# change state_minus into state_plus
print(state_minus_1)
```

```python
np.dot(state_minus_1.T, state_minus)
```

```python
np.dot(state_0_1.T, state_1)
```

```python
from qiskit.quantum_info import Statevector# for quantum state vector

state_2 = Statevector([(1 + 2.0j) / 3, -2 / 3])#make a new quantum state by \bra{0} and \bar{1}
print(state_2.is_valid())# double checke the probability amplitude is equal to 1!
state_2.draw('latex')
```

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(3)
# h|0> -> |+>
qc.h(0)
qc.i(1)
qc.cnot(0, 1)
#The i gate preserves the quantum state's current condition without any change
qc.i(2)
qc.i(2)
qc.draw('mpl')
```
